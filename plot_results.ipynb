# Recreating graphs and evaluating them
#### All graphs are created from the full dataset, the train-test split has not been done here.

```python
nodes, edgelist = load_data_from_csv(edgelist_path='data/edgelist.csv', node_path='data/node_data.csv')
```

```python
G = create_graph_with_attributes(nodes, edgelist)
```

```python
plotting.plot_graph_features(G,'degree')
```

The log-log plot shows that the degree distribution is not scale-free as there is a visible break in the line around x=3. The closest distribution is maybe the exponential distribution that could characterise this network.

```python
plotting.plot_graph_features(G,'agedist')
```

From the age distribution chart, we can see the following important data:

* In the dataset, the young people are overrepresented. with the highest number for ages between 20-25. The reason can be that the dataset is from an online platform where younger generations are more likely to be present.
* The maximum age that are represented in the dataset to a visible extent is those around 55. Above that age, very few users are in the dataset.
* There are in general more male users than females in the 20-40 age groups and a bit more females in the younger generation and above 40.

```python
plotting.plot_graph_features(G,'descriptive')
```

Comparing the degree centrality with the chart in the article we can see that there is a significant peak around age 20-25, that is similar to the phone call network presented by Dong et al. After that age the avg degree decreases to around 4-5, just as in the article. One difference is that in our dataset there is no peal around the age of 50, but the reason can be that there is not too much data for people around that age. This phenomena calusing also the high deviation after age 60. There is not enough data. In the article, there was a slight contrast between males and females regarding the degree distribution but not in our dataset.

For neighbor connectivity chart we can see that there is also a peak around age 20-25, similarly to the article. A second bump also appears between 40-60 in both datasets. Interestingly, in our plot, females have higher neighbor connectivity while in the article males were a bit above on the chart. This chart has also no added value after the age of 60 due to low number of users in that age.

In the first two charts the peaks are much higher than in the article, which can be understood by looking at the source of the data. While one is constructed from phone calls made, this one is from friendships on a social network. Clearly it is easier to send a friend request to someone than call them several times, so this can be the key why our dataset have higher average degrees in general.

Looking at the charts in the second row, we can spot that the clustering coefficient slightly decreases till the age of 40, then rises a bit, while in the article it rises over the time. The embeddedness takes a peak around age 20 then plummets, while in the article it rises constantly. This difference can also lie in different source of data (making a phone call is an activity that has to be carried out regularly but making a friendship on the social media is a one-time action). That is why making sociological conclusions is hard from these two subplots.


```python
plotting.plot_heatmap_demog_dist(nodes, edgelist)
```

The chart above shows the distribution of friends's ages for male and female users. On the y axis, males' age is shown negative to add a new dimension to the plots. We can see a significant 45 line for females (both for female and male friends, but for younger ages it is a bit more intense for the same gender friends. From these we can make the conclusion that females in general tend to make friendships from their own age regardless of the gender of the friend.

The second subplot, presenting the friendships of male users tells us that male users in general also tend to make friends in the similar age as they are, also with a higher proportion of friends from the same gender for younger ages. Here the line though more intense for the opposite gender friends from the same age group for older ages. 

In both subplotst here are also some higher intensity areas for older ages to younger ages on the right side of the plot. These users can be grandparents registering to the platform and making friendships with their children, grandchildren. 

```python
plotting.plot_age_group_proportion(nodes, edgelist)
```

From the graph above we can see that young users (males, females as well) tend to have mostly friends from their age group [-5, +5] range but this proportion decreases for older users. People from older generations have mostly friends from younger generations [-30, -20] range, while the younger generation having older friends is not visible. The reason is that younger users have more friends in general from which the older users make just a small amount, while older generation users have less friends, from which the younger friends take higher proportion. These people have more friends from youngers than from their age group, that can be argued by the parent-children relationships appearing on the social media. The same patterns are visible for male and female users as well. Both genders having more female friends from their age group in their early ages as it was already presented in the age distribution chart.

All these findings are in line with the graphs presented in the article.

From the triad heatmaps we can also obtain important data about the composition of the friend groups. The logarhitmic charts give more detailed information so we'll base our finding on them. The other two different compositions are closer to MMM friend groups.

Comparing to the article's charts, in our dataset there aren't many groups where one member is from an older generation but that can also be an effect of lack of data on the olders.

The charts above represent the age in different pairs of friendships. From the logarithmic chart we can discover more patterns so in the following we'll refer to that one. The frequent part being among the younger generation can be also a cause of these generation using the platform mostly so the absolute number of connections between them is higher compared to others.
